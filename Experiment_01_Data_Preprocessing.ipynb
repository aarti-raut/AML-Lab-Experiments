# Experiment 1: Data Preprocessing
**Dataset:** Iris Dataset (UCI ML Repository)
**Aim:** Apply data preprocessing techniques to make data suitable for machine learning.

```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split

# Step 2: Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print('=== First 5 rows of dataset ===')
print(df.head())
```

```python
# Step 3: Check basic info and missing values
print('=== Dataset Shape ===')
print(df.shape)

print('\n=== Dataset Info ===')
print(df.info())

print('\n=== Missing Values ===')
print(df.isnull().sum())

print('\n=== Basic Statistics ===')
print(df.describe())
```

```python
# Step 4: Introduce some missing values artificially to practice handling
df_copy = df.copy()
np.random.seed(42)
missing_indices = np.random.choice(df_copy.index, size=10, replace=False)
df_copy.loc[missing_indices, 'sepal length (cm)'] = np.nan

print('Missing values after introduction:')
print(df_copy.isnull().sum())

# Fill missing values with mean
df_copy['sepal length (cm)'].fillna(df_copy['sepal length (cm)'].mean(), inplace=True)
print('\nMissing values after filling with mean:')
print(df_copy.isnull().sum())
```

```python
# Step 5: Label Encoding (converting text labels to numbers)
le = LabelEncoder()
df['species_encoded'] = le.fit_transform(df['species'])
print('=== Label Encoding ===')
print(df[['species', 'species_encoded']].drop_duplicates())
```
