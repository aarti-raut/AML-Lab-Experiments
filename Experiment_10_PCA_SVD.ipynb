# Experiment 10: PCA and SVD
**Dataset:** Iris Dataset (UCI ML Repository)
**Aim:** Apply PCA and SVD on Iris dataset obtained from UCI ML repository.

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target
print('Dataset: Iris')
print(f'Original Shape: {X.shape} (150 samples, 4 features)')
```

```python
# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# === PCA ===
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print('=== PCA Results ===')
print(f'Reduced Shape: {X_pca.shape} (150 samples, 2 principal components)')
print(f'Explained Variance Ratio: {pca.explained_variance_ratio_}')
print(f'Total Variance Explained: {sum(pca.explained_variance_ratio_)*100:.2f}%')
```

```python
# Visualize PCA
plt.figure(figsize=(8, 5))
colors = ['red', 'green', 'blue']
for i, name in enumerate(iris.target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1],
                c=colors[i], label=name, alpha=0.7)
plt.title('PCA - Iris Dataset (4D → 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# === SVD ===
svd = TruncatedSVD(n_components=2, random_state=42)
X_svd = svd.fit_transform(X_scaled)

print('=== SVD Results ===')
print(f'Reduced Shape: {X_svd.shape}')
print(f'Explained Variance Ratio: {svd.explained_variance_ratio_}')
print(f'Total Variance Explained: {sum(svd.explained_variance_ratio_)*100:.2f}%')

# Visualize SVD
plt.figure(figsize=(8, 5))
for i, name in enumerate(iris.target_names):
    plt.scatter(X_svd[y == i, 0], X_svd[y == i, 1],
                c=colors[i], label=name, alpha=0.7)
plt.title('SVD - Iris Dataset (4D → 2D)')
plt.xlabel('SVD Component 1')
plt.ylabel('SVD Component 2')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# PCA Scree Plot - how many components to keep
pca_full = PCA()
pca_full.fit(X_scaled)

plt.figure(figsize=(7, 4))
plt.bar(range(1, 5), pca_full.explained_variance_ratio_ * 100, color='steelblue')
plt.plot(range(1, 5), np.cumsum(pca_full.explained_variance_ratio_ * 100),
         marker='o', color='red', label='Cumulative')
plt.title('PCA Scree Plot - Variance Explained per Component')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained (%)')
plt.legend()
plt.tight_layout()
plt.show()

print('\nConclusion: First 2 PCs capture most of the variance!')
```
