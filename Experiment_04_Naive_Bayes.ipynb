# Experiment 4: Naive Bayes Classifier
**Dataset:** Iris Dataset (UCI ML Repository)
**Aim:** Train the system using Iris dataset and determine accuracy using Naive Bayes.

```python
from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target
print('Dataset: Iris')
print(f'Samples: {X.shape[0]}, Features: {X.shape[1]}')
```

```python
# Preprocess and split
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)

# Train Gaussian Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
print('Naive Bayes model trained!')
print(f'Classes: {iris.target_names}')
```

```python
# Predict and evaluate
y_pred = nb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f'=== Model Accuracy: {accuracy * 100:.2f}% ===')
print('\n=== Classification Report ===')
print(classification_report(y_test, y_pred, target_names=iris.target_names))
print('=== Confusion Matrix ===')
print(confusion_matrix(y_test, y_pred))
```

```python
# Show class probabilities for first 5 test samples
print('=== Prediction Probabilities (first 5 test samples) ===')
probs = nb_model.predict_proba(X_test[:5])
import pandas as pd
prob_df = pd.DataFrame(probs, columns=iris.target_names)
prob_df['Predicted'] = [iris.target_names[p] for p in nb_model.predict(X_test[:5])]
prob_df['Actual'] = [iris.target_names[a] for a in y_test[:5]]
print(prob_df.round(3))
```
